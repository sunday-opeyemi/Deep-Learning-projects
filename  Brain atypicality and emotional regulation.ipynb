```python
# To extract all the file in the sub-folders and convert to dataset
import os
import nibabel as nib
import pandas as pd
import json
from nilearn import image, masking
from nilearn.input_data import NiftiMasker

# Define the root directory
root_dir = 'C:\\Part 1_Brain Atypicality and Emotional Regulation'
# Initialize a list to store the results
results = []

# Loop through each subfolder in the root directory
for subject_folder in os.listdir(root_dir):
    subject_path = os.path.join(root_dir, subject_folder)
    
    if os.path.isdir(subject_path):
        # Define paths to anat and func folders
        anat_dir = os.path.join(subject_path, 'anat')
        func_dir = os.path.join(subject_path, 'func')
        
        # Load the structural MRI (anat folder)
        if os.path.exists(anat_dir):
            t1w_file = [f for f in os.listdir(anat_dir) if f.endswith('_T1w.nii.gz')]
            if t1w_file:
                struct_img = nib.load(os.path.join(anat_dir, t1w_file[0]))
            
            # Load the fMRI scans and corresponding files (func folder)
            bold_files = [f for f in os.listdir(func_dir) if f.endswith('_bold.nii.gz')]
            events_files = [f for f in os.listdir(func_dir) if f.endswith('_events.tsv')]
            physio_files = [f for f in os.listdir(func_dir) if f.endswith('_physio.tsv.gz')]
            json_files = [f for f in os.listdir(func_dir) if f.endswith('.json')]

            for bold_file in bold_files:
                task_name = bold_file.split('_')[1]  # Assuming task name is in the filename
    
                fmri_img = nib.load(os.path.join(func_dir, bold_file))
    
                # Find corresponding events and physio files
                events_file = [f for f in events_files if task_name in f]
                physio_file = [f for f in physio_files if task_name in f]
                bold_json = [f for f in json_files if 'bold' in f and task_name in f]
                physio_json = [f for f in json_files if 'physio' in f and task_name in f]
    
                if events_file:
                    events_data = pd.read_csv(os.path.join(func_dir, events_file[0]), sep='\t')
                if physio_file:
                    physio_data = pd.read_csv(os.path.join(func_dir, physio_file[0]), sep='\t')
                if bold_json:
                    with open(os.path.join(func_dir, bold_json[0])) as f:
                        metadata_bold = json.load(f)
                if physio_json:
                    with open(os.path.join(func_dir, physio_json[0])) as f:
                        metadata_physio = json.load(f)
    
                # Preprocess fMRI data (e.g., smoothing, masking)
                masker = NiftiMasker(standardize=True)
                fmri_data = masker.fit_transform(fmri_img)
    
                # Conduct analysis on brain atypicality and emotional regulation
                mean_activation = fmri_data.mean(axis=0)
    
                # Save or process the data further for your specific analysis
                output_file = os.path.join(subject_path, f'mean_activation_{task_name}.csv')
                pd.DataFrame(mean_activation).to_csv(output_file, index=False)

                # Store the results in a dictionary
                result = {
                    'subject': subject_folder,
                    'task': task_name,
                    'mean_activation': mean_activation
                }
                results.append(result)

                # You can also save the processed NIfTI data if needed
                # processed_img = masker.inverse_transform(mean_activation)
                # processed_img.to_filename(os.path.join(subject_path, f'processed_{task_name}_bold.nii.gz'))
```

```python

# Convert the results into a DataFrame
results_df = pd.DataFrame(results)

# Save the DataFrame to a CSV file
output_file = os.path.join(root_dir, 'mean_activation_results.csv')
results_df.to_csv(output_file, index=False)
```
