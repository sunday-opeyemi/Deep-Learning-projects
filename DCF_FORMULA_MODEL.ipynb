```python

import numpy as np

# Known x-values
known_x = np.array([1, 2, 3, 4])

# Known y-values
y_values=[]
while True:
    y_values_input=int(input("supply the knwn fcf for previous years: "))
    y_values.append(y_values_input)
    user = input('press enter to input another value or 1 to stop: ')
    if user == '1':
        break

#print( y_values)

known_y = np.array(y_values)
#known_y = np.array([178710000, 100747000 , 14946000 ,35416000])


# Value of x for which you want to forecast a value
x_forecast = 5

# Perform linear regression to find the coefficients of the linear relationship
coefficients = np.polyfit(known_x, known_y, 1)

# Extract the slope and intercept from the coefficients
slope = coefficients[0]
intercept = coefficients[1]

# Use the linear equation to forecast the value of y for the given x_forecast
y_forecast = slope * x_forecast + intercept
FCF=abs(y_forecast)

print("Forecasted value of y for x =", x_forecast, "is:", FCF)
#slope


 
```

```python
#to calculate TV(Terminal values)
WACC=6.8/100
growth_rate=7/100
TV= (FCF*(1+growth_rate))/(WACC-growth_rate)
TV=abs(TV)
TV
```

```python
#to calculate the discount factor for each period

#discount factor for period 1
DF1=float(1)/(float(1)+float(WACC))**1
DF1



#discouunt factor for period 2
DF2=float(1)/(float(1)+float(WACC))**2
DF2


#discouunt factor for period 3
DF3=float(1)/(float(1)+float(WACC))**3
DF3

#discouunt factor for period 4
DF4=float(1)/(float(1)+float(WACC))**4
DF4


#discouunt factor for period 2
DF5=float(1)/(float(1)+float(WACC))**5
DF5

DF1
```

```python
#to find the present value of free cash flow PVofFCF

#PV of FCF FOR period 1
PV_FCF1 = y_values[0] * DF1
PV_FCF1 


#PV of FCF FOR period 2
PV_FCF2 = y_values[1] * DF2
PV_FCF2


#PV of FCF FOR period 3
PV_FCF3 = y_values[2] * DF3
PV_FCF3


#PV of FCF FOR period 4
PV_FCF4 = y_values[3] * DF4
PV_FCF4



#PV of FCF FOR period 5
PV_FCF5 = FCF * DF5
PV_FCF5 

PV_FCF1 
```

```python
#Present value of Terminal Value PV of TV
PV_ofTV=TV * DF5
PV_ofTV
```
